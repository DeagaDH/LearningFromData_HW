## Alternative c:
First, let us manipulate the expression. The given expression for the Parrondo and Van den Broak bound is:

\begin{equation}
\epsilon \leq \sqrt{\frac{1}{N} \Big( 2 \epsilon + \ln \frac{6 m_H(2N)}{\delta} \Big)}
\end{equation}

Squaring both sides and going some algebraic manipulation, we can get a second degree polynomial on $\epsilon$:

\begin{equation}
\epsilon^2 \leq \frac{1}{N} \Big( 2 \epsilon + \ln \frac{6 m_H(2N)}{\delta} \Big) \\
\epsilon^2 \leq \frac{2}{N}\epsilon + \frac{1}{N} \ln\frac{6 m_H(2N)}{\delta} \\
\epsilon^2 + \underbrace{\Big( -\frac{2}{N}\Big)}_b \epsilon + \underbrace{\Big( - \frac{1}{N} \ln\frac{6 m_H(2N)}{\delta}\Big)}_c \leq 0 \\
\epsilon^2 + b \epsilon + c \leq 0
\end{equation}

We can then use the quadratic formula to solve this expression: 

\begin{equation}
\epsilon_{1,2} \leq \frac{-b \pm \sqrt{b^2 - 4c}}{2} \text{ , }
\end{equation}

with $b = -\frac{2}{N}$ and $c = - \frac{1}{N} \ln\frac{6 m_H(2N)}{\delta}$

```python
#Run experiment N times, using a_avg from Question 4. Don't modify a_avg nor xlim
N=10000

#Star bias and var at 0
bias=0
var=0

#Initialize a line0 object with angular coefficient a_avg
learned_line = line0(angular=a_avg)

for i in range(N):
    # Generate random point on the sine wave
    p = random_point_sine(x_lim) #x_lim from question 4

    # Bias on this point: squared error between the value of f(x) (p[1], the sine) and the value
    f_x = p[1]
    g_x = learned_line.get_y(p[0])
    bias += (f_x - g_x)**2

    #Variance: make another y = ax line and compare it to our given line
    p1 = random_point_sine(x_lim)
    p2 = random_point_sine(x_lim)
    another_line=line0(p1=p1,p2=p2)

    #Now compare
    gD_x = another_line.get_y(p[0])
    var += (gD_x - g_x)**2

#Turn into averages
bias = bias/N
var = var/N

#Print results:
print(f'Average bias: {round(bias,2)}')
print(f'Average variance: {round(var,2)}')
```

```python
from scipy.optimize import curve_fit
from hw4_func import *

############################################
###                PART 1
###    MAKE FITS FOR ALL ALTERNATIVES   
############################################

#Run experiment N times, then get the average
N = 10000

#Initialize average value of a, a_avg
a_avg = 0

#Limit x values to this range
x_lim = [-1,1]

#Initialize lists for the average coefficients of all alternatives.
#Use lists for all, even the ones with just one coefficient, for consistency
a_avg=[0]
b_avg=[0]
c_avg=[0,0]
d_avg=[0]
e_avg=[0,0]

# Use a N datasets to fit all alternatives
for i in range(N):
    
    #Display current experiment in console, every 500 experiments
    if (i+1 == 1) or ((i+1) % 500 ==0):
        print(f'Running fit number {i+1}...')

    #Create two random points on the sinusoid
    #Return values as two lists, x values and y values, to use with SciPy
    x,y = random_point_sine_list(size=2,xlim=x_lim)

    #Fit the functions from all alternatives using SciPy
    a,_ = curve_fit(q7_func_a,x,y)
    b,_ = curve_fit(q7_func_b,x,y)
    c,_ = curve_fit(q7_func_c,x,y)
    d,_ = curve_fit(q7_func_d,x,y)
    e,_ = curve_fit(q7_func_e,x,y)

    #Add coefficients to the average values
    a_avg[0] += a[0]
    b_avg[0] += b[0]
    c_avg[0] += c[0]
    c_avg[1] += c[1]
    d_avg[0] += d[0]
    e_avg[0] += e[0]
    e_avg[1] += e[1]
    
#Turn all coefficients into averages
for i in range(2):
    
    if (i==0):
        a_avg[i] /= N
        b_avg[i] /= N
        d_avg[i] /= N

    c_avg[i] /= N
    e_avg[i] /= N


############################################
###                PART 2
###    BIAS/VARIANCE FOR ALL FITS 
############################################

#Run experiment N times
N=10000

#Star bias and var at 0. Each element in list is the bias/var for one alternative
bias=[0,0,0,0,0]
var=[0,0,0,0,0]

for i in range(N):

    #Display current experiment in console, every 500 experiments
    if (i+1 == 1) or ((i+1) % 500 ==0):
        print(f'Calculating bias/var on cycle number {i+1}...')

    # Generate random point on the sine wave
    x,y = random_point_sine_list(size=1,xlim=x_lim) #x_lim from question 4
    x=x[0]
    y=y[0]

    # Bias on this point: squared error between the value of f(x) (p[1], the sine) and the value
    f_x = y
    g_x = [q7_func_a(x,a_avg[0]),
            q7_func_b(x,b_avg[0]),
            q7_func_c(x,c_avg[0],c_avg[1]),
            q7_func_d(x,d_avg[0]),
            q7_func_e(x,e_avg[0],e_avg[1])]

    for i in range(5):
        bias[i] += (f_x - g_x[i])**2

    #Variance: make another fit to two random points and compare to average fits
    x2,y2 = random_point_sine_list(size=2,xlim=x_lim) #x_lim from question 4

    a,_ = curve_fit(q7_func_a,x2,y2)
    b,_ = curve_fit(q7_func_b,x2,y2)
    c,_ = curve_fit(q7_func_c,x2,y2)
    d,_ = curve_fit(q7_func_d,x2,y2)
    e,_ = curve_fit(q7_func_e,x2,y2)

    #Now evauate gD_x with the fit from the current dataset
    gD_x = [q7_func_a(x,a[0]),
            q7_func_b(x,b[0]),
            q7_func_c(x,c[0],c[1]),
            q7_func_d(x,d[0]),
            q7_func_e(x,e[0],e[1])]

    #And get the variance for each alternative
    for i in range(5):
        var[i] += (gD_x[i] - g_x[i])**2


#Turn bias and var into averages
for i in range(5):
    bias[i] /= N
    var[i] /= N

#Print results:
alternatives=['a','b','c','d','e']

for alt,bias_alt,var_alt in zip(alternatives,bias,var):
    print()
    print(f'ALTERNATIVE {alt}')
    print(f'Average bias: {round(bias_alt,2)}')
    print(f'Average variance: {round(var_alt,2)}')
    print(f'Average error: {round(bias_alt+var_alt,2)}')
```

# QUESTION 8

From the problem statement, we know that:

\begin{align}
q &\geq 1 quad \text{and $q$ is an integer}\\
{M \choose m} &= 0 \quad \text{, for} \quad m > M \\
m_H(1) &= 2 \\
m_H(N+1) &= 2 m_H(N) + {N \choose q}
\end{align}

We can rewrite the last equation in the equivalent form:

\begin{equation}
m_H(N) = 2 m_H(N-1) + {N-1 \choose q} \text{ , }
\end{equation}

which is a function of $N$ rather than $N+1$. let us first focus on the first term on the right hand side of the above equation, $2 m_H(N-1)$. We can only evalute this term recursively, until we reach the base case $m_H(1) = 2$, such as shown below:

\begin{align}
2 m_H(N-1) = 2\underbrace{(2m_H(N-2))}_{m_H(N-1)} = 2\underbrace{(2(2m_H(N-3)))}_{m_H(N-2)} = .... 2^{N-1}(m_H(1)) = 2^N
\end{align}

As we can see, the recursion leads us to the result $2 m_H(N-1) = 2^N$ as we'll do the multiplication by 2 $N-1$ times, with an additional multiplication by 2 as the base case is $m_H(1) = 2$.

We also know, from the definition of the VC dimension, that the VC dimension is the maximum $N$ we can shatter in a data set. That is equivalent to saying that the VC dimension is the maximum N for which we have:

\begin{equation}
m_H(N) \equiv 2^N \text{ . }
\end{equation}

Since we have shown that $2 m_H(N-1) = 2^N$, we can rewrite the growth function from the problem statement as:

\begin{equation}
m_H(N) = \underbrace{2^N}_{2 m_H(N-1)} - {N-1 \choose q}
\end{equation}

This expression is identically equal to $2^N$ when ${N-1 \choose q} = 0$. We also know that ${N-1 \choose q} = 0$ if $q > N-1$. Since $q$ is an integer, the inequality will hold up until $N = q$; for higher values of $N$ (ie $N > q+1$), we will have ${N-1 \choose q} > 0$ and $m_H(N) < 2^N$. Therefore, **we have that $VC_{dim} = q$ (Alternative c)**.
